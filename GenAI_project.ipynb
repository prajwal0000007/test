```python
%%writefile app.py
import streamlit as st
import torch
from diffusers import StableDiffusionPipeline

st.set_page_config(page_title="AI Interior Design Generator", layout="centered")

st.title("üè° AI Interior Design Generator")
st.write("Generate interior designs using Generative AI")

room = st.selectbox("Select Room Type", ["Bedroom", "Living Room", "Home Office"])
style = st.selectbox("Select Style", ["Modern", "Minimalist", "Luxury", "Boho"])
lighting = st.selectbox("Lighting Preference", ["Soft", "Warm", "Natural"])

prompt = f"""
{style} {room} interior design,
{lighting} lighting,
realistic, high resolution,
interior design photography
"""

@st.cache_resource
def load_model():
    pipe = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16
    )
    return pipe.to("cuda")

pipe = load_model()

if st.button("‚ú® Generate Design"):
    with st.spinner("Designing your room..."):
        image = pipe(prompt).images[0]
        st.image(image, caption="Generated Interior Design")
```
